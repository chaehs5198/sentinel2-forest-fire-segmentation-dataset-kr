```python
import os
from PIL import Image
import torch
```

```python
import rasterio
```

```python
# Set the directory where your mask files are stored
mask_directory = r'D:\deep_learning\Sentinel_forestfire\MY_TEST_SENTINEL_IMAGERY_DATA\imagery_data_complete_0828\data\mask'

# Initialize variables to keep track of the counts
total_positive_pixels = 0
total_negative_pixels = 0

# Iterate through the GeoTIFF mask files in the directory
for filename in os.listdir(mask_directory):
    with rasterio.open(os.path.join(mask_directory, filename)) as src:
        # Read the GeoTIFF mask as a numpy array
        mask_array = src.read(1)
        
        # Convert the numpy array to a PyTorch tensor
        mask_tensor = torch.tensor(mask_array)
        
        # Count the number of positive (1) values in the tensor
        positive_count = torch.sum(mask_tensor == 1).item()
        
        # Count the number of negative (0) values in the tensor
        negative_count = torch.sum(mask_tensor == 0).item()
        
        # Update the total counts
        total_positive_pixels += positive_count
        total_negative_pixels += negative_count

# Calculate the proportions
total_pixels = total_positive_pixels + total_negative_pixels
positive_proportion = total_positive_pixels / total_pixels
negative_proportion = total_negative_pixels / total_pixels

# Print the proportions
print("Proportion of positive pixels:", positive_proportion)
print("Proportion of negative pixels:", negative_proportion)
```
