```python
new_df = new_df.sample(frac=1, random_state=6183)
```

```python
train_ratio, val_ratio, test_ratio = 0.6, 0.2, 0.2
```

```python
new_df['dataset'] = 'train'
```

```python
grouped = new_df.groupby('규모구분(등급)')
```

```python
for _, group in grouped:
    num_samples = len(group)
    num_train = int(train_ratio * num_samples)
    num_val = int(val_ratio * num_samples)
    
    indices = np.arange(num_samples)
    np.random.shuffle(indices)
    
    train_indices = indices[:num_train]
    val_indices = indices[num_train:num_train + num_val]
    
    new_df.loc[group.index[train_indices], 'dataset'] = 'train'
    new_df.loc[group.index[val_indices], 'dataset'] = 'val'
    new_df.loc[group.index[num_train + num_val:], 'dataset'] = 'test'
```

```python
image_source_folder = 'image'
mask_source_folder = 'mask'
```

```python
train_dest_folder = '../train'
test_dest_folder = '../test'
val_dest_folder = '../val'
```

```python
for index, row in tqdm(new_df2.iterrows(), total = new_df2.shape[0]):
    fname = row['filename']
    mname = row['maskfilename']
    dataset = row['dataset']
    
    source_image_path = os.path.join(image_source_folder, fname)
    source_mask_path = os.path.join(mask_source_folder, mname)
    
    if dataset == 'train':
        destination_image_folder = train_dest_folder
        destination_mask_folder = train_dest_folder
    elif dataset =='val':
        destination_image_folder = val_dest_folder
        destination_mask_folder = val_dest_folder
    elif dataset =='test':
        destination_image_folder = test_dest_folder
        destination_mask_folder = test_dest_folder
    else:
        print(f"Invalid dataset label: {dataset}")
        continue
        
    destination_image_path = os.path.join(destination_image_folder, 'image', fname)
    destination_mask_path = os.path.join(destination_mask_folder, 'mask' ,mname)
    
    #shutil.copy(source_image_path, destination_image_path)
    shutil.copy(source_mask_path, destination_mask_path)
    
print('FIle copying completed.')
    
    
    
    
```
