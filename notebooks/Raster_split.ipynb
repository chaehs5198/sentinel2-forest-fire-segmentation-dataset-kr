```python
import rasterio
from rasterio.windows import Window
import os
from glob import glob


def split_image(input_path, output_path, tile_size=256):
    with rasterio.open(input_path) as src:
        width, height = src.width, src.height
        meta = src.meta.copy()
        meta.update({
            'driver': 'GTiff',
            'height': tile_size,
            'width': tile_size,
            #'count': 6,
            #'transform': rasterio.windows.transform(Window(col_off=0, row_off=0, width=tile_size, height=tile_size), src.transform),
        })
        for col in range(0, width, tile_size):
            for row in range(0, height, tile_size):
                window = Window(col_off=col, row_off=row, width=tile_size, height=tile_size)
                tile = src.read(window=window)
                transform = rasterio.windows.transform(window, src.transform)
                meta.update({'transform': transform})
                outpath = f"{output_path}{input_path[:-4]}_{col}_{row}.tif"
                with rasterio.open(outpath, 'w', **meta) as dst:
                    dst.write(tile)
```
