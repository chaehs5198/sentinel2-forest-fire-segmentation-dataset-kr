```python
import os
from glob import glob
import rasterio
from rasterio.enums import Resampling
```

```python
os.chdir(r'D:\deep_learning\Sentinel_forestfire\MY_TEST_SENTINEL_IMAGERY_DATA\imagery_data_complete_0828\erdas_isodata')
# Define the input and output directories
input_dir = "tiffs"
output_dir = "imgs"
```

```python
tiffs = glob(os.path.join(input_dir, '*.tif'))
```

```python
tiffs
```

```python
for tiff in tiffs:
    input_tiff_path = tiff
    output_img_path = os.path.join(output_dir, os.path.splitext(os.path.basename(tiff))[0] + ".img")
    
    with rasterio.open(input_tiff_path) as src:
        # Copy the CRS
        num_bands = src.count

        # Copy the band count and metadata
        num_bands = src.count
        profile = src.profile.copy()

        # Resample if needed (e.g., to change resolution)
        # src = src.read(
        #     out_shape=(src.count, src.height, src.width),
        #     resampling=Resampling.bilinear
        # )
        
        profile.update(driver='HFA')

        # Write to Imagine format
        with rasterio.Env():
            with rasterio.open(output_img_path, 'w', **profile) as dst:
                for i in range(1, num_bands + 1):
                    dst.write(src.read(i), indexes=i)

        print(f"Converted {tiff} to Imagine format.")

print("Conversion completed.")
    
    
    
    
```
